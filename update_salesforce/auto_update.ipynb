This notebook is to auto update the information of TI Eligibility in Salesforce TIP Account

```python
# import necessary libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read accounts information from the file
accounts = pd.read_csv('Account.csv')
accounts.head()
```

```python
# source of truth for the account information
source_of_truth = pd.read_csv('source_of_truth.csv')
source_of_truth.head()
```

```python
import pandas as pd

# Load the source of truth data
source_of_truth = pd.read_csv('source_of_truth.csv')

def concat_columns(group):
    y1_eligibility = f"Y1 - {group['AOC']}: {group['Y1 Final Status (Post-Email Check) (AOC Level)']};"
    y2_eligibility = f"Y2 - {group['AOC']}: {group['Y2 Final Status (Post-Email Check) (AOC Level)']}"
    return y1_eligibility + y2_eligibility

# Group by 'Organization Name' and apply the concatenation function
source_of_truth['TI Eligibility'] = source_of_truth.apply(lambda x: concat_columns(x), axis=1)

# Display the updated DataFrame
source_of_truth.head()
```

```python
# Group by 'Organization Name' and concatenate the 'TI Eligibility' values
grouped = source_of_truth.groupby('Organization Name')['TI Eligibility'].apply(lambda x: ';'.join(x)).reset_index()

# Display the updated DataFrame
grouped.head()
```

```python
grouped.to_csv('grouped.csv', index=False)
```

```python
# map the ti eligibility to the accounts
accounts['TI Eligibility Y2'] = accounts['Name'].map(grouped.set_index('Organization Name')['TI Eligibility'])
accounts
```

```python
accounts.to_csv('updated_accounts.csv', index=False)
```
